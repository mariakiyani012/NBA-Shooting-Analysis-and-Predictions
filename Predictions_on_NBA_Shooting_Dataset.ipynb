## Exploratory Data Analysis (EDA)
The dataset is not huge, we can easily get to know it and decide how to tackle some interesting questions. From the introductory code above we know:
* `SHOOTER` and `DEFENDER` are categorical variables taking 4 values (the four player names)
* `RANGE` is also categorical containing 6 bins
* `SCORE` is binary
* `X` and `Y` are numerical continuous variables  

First of all, let's ensure there are no nulls:

```python
df.isnull().sum()
```

```python
# Let's check the values `X` and `Y` take:
df.describe()
```

There is nothing suspicious there, no outliers or other non-ordinary values. What might come handy is a new variable with the exact range the shot was taken from, instead of having the range split in bins, like it does in `RANGE`. It is easy to caluclate the variable `RANGE_num` from the `X` and `Y` coordinates; then, we compare the exact numerical values with the bins to validate the new variable.

```python
# calculate the new variable
df['RANGE_num'] = np.sqrt(df.X**2 + df.Y**2)
# validate the calculation by chekcing 10 random samples:
df[['RANGE','RANGE_num']].sample(10)
```

We can see that the `RANGE_num` of each row is contained in the corresponding bin, so the calculation is correct. The `RANGE` variable is of type Object, and its values do not sort properly: the bin (5,9) appears last. This is counter-intuitve when displaying visuals, so we will convert the variable to ordered categorical with the correct ordering

```python
df['RANGE'] = df.RANGE.astype("category")
print("The variable 'RANGE' is not sorted properly:\n", df.RANGE.cat.categories)
df['RANGE'] = df.RANGE.cat.reorder_categories(['(0, 4)', '(5, 9)', '(10, 14)', '(15, 19)', '(20, 24)', '(25, 29)'], ordered=True)
print("Now it sorts properly:\n", df.RANGE.cat.categories)
```

### What is the best shooting position for each of the four players?

These scatter plots were not very helpful, there is no obvious clustering of successful shots.  Let's do some feature engineering by separating the field in three regions, the area within 9 feet of the basket, the area from 9 feet to the 3-pointer line (at 23.9 feet), and the area beyond the 3-pointer; a new variable `RANGE_region` will hold that info, and we validate the new variable by sampling 10 shots. ## Predict the outcome of a shot
Can we train a model to predict if a shot will be `MADE` or `MISSED`? Let's try with several models and find the one that predicts with better accuracy. This is a classification problem, so let's see if the outcome classes `MADE` and `MISSED` are balanced:

```python
# Is the outcome balanced or not?
df.SCORE.value_counts(normalize=True)
```

Since the outcomes are balanced (54%:46%) we can prepare the data set for training. We will use the two numeric features: `X`, `Y`. We don't need the `RANGE_region` and `RANGE_num` features since we have the `RANGE`, and testing showed that it is much more informative. We will also use the `SHOOTER` and `DEFENDER` for training by binary-encoding the player names. The ground truth will be the `SCORE` replaced with 0's for `MISSED`  and 1's for `MADE`.

The distribution of `X` is normal and `Y` is normal-like, since there is no negative values.

```python
sns.histplot(df[['X','Y']], bins=40,alpha=.5)
```

So, let's construct the X and y sets for the process of training & validating a model (not to be confused with `X` and `Y` variables that are the coordinates of the shoot). The sets are then split into `train` and `test` sets, for training an dvalidating respectively. A *ColumnTransformer* will be used to preprocess the data: the numerical features will be scaled with *MinMaxScaler* and the categorical will be one-hot encoded. This transformer will be used in a pipeline later.

#### Is the data seperable?
We can inspect the set to see if there are obvious clusters and how separable the data is. The data dimensions will be reduced  with the t-SNE algorithm to two features and visualized in a scatter plot.


```python
NUMERIC = ['X', 'Y', ]
CATEGORICAL = ['SHOOTER','DEFENDER', 'RANGE']

X = df[NUMERIC + CATEGORICAL]
y = df['SCORE'].replace(['MADE','MISSED'],[1,0])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=23)

preprocess = ColumnTransformer( transformers=[
    ('onehot',OneHotEncoder(), CATEGORICAL),
    ('scale', MinMaxScaler(), NUMERIC),
])
```

The visualization shows that the dataset is not linearly separable, so classification is difficult to achieve. In this case the dataset can be projected to a higher dimension, which might be more convenient for classification. We can use *PolynomialFeatures* transformer to do this.

#### Setting up the pipelines
Pipelines will be set up for each model we will try: Random Forest and SVC are good models for binary classification, Logistic Regression is included for benchmarking and a Multi-Level Perceptron (MLP), too

```python
#polynomial transformer
polynomial = PolynomialFeatures(2, interaction_only=False, )
#set up pipelines with preprocessing, polynomial transformation & estimators
pipelines = {
    'lr' : make_pipeline(preprocess, polynomial, LogisticRegression()),
    'rf' : make_pipeline(preprocess, polynomial, RandomForestClassifier()),
    'svc': make_pipeline(preprocess, polynomial, SVC()),
    'mlp': make_pipeline(preprocess, polynomial, MLPClassifier(max_iter=1000))
}
```

#### Evaluating the simple models
The *roc-curve* shows how well the model performs: The more area under the curve, the better the model. We can plot the curve for each of the four models we trained, and show its *accuracy* (ACC) and *Area-Under-the-Curve* (AUC). The visual shows that ***svc* is the best of the naive models**, with the highest AUC and ACC.

```python
_, ax = plt.subplots()

plt.plot([0, 1], [0, 1], linestyle='--', c='black')

# Loop over the estimators in the pipelines dictionary
for name, estimator in pipelines.items():
    # Fit the model and predict on the test set
    estimator.fit(X_train, y_train)
    y_pred = estimator.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    # Plot the ROC curve using RocCurveDisplay
    RocCurveDisplay.from_estimator(estimator, X_test, y_test, name=f"{name} (ACC= {acc:.2f})", ax=ax)

plt.show()
```

#### Hyperparameter fine-tuning
Now that we have an idea of the bare minimum of model performance we can perform a Grid Search to fine-tune the hyperparameters of each model. The score we want to maximize is *accuracy*.

```python
# Assuming 'pipelines' is a dictionary containing your models
# Example: pipelines = {'lr': logistic_regression_pipeline, 'rf': random_forest_pipeline, ...}

# Hyperparameter grids
lr_hyperparameters = {
    'logisticregression__C': [0.001, 0.005, 0.01, 0.05, 0.1],
    'logisticregression__penalty':['l1','l2'],
    'logisticregression__tol':[0.0001, 0.0003, 0.001],
    'logisticregression__max_iter':[200,500,100]
}

rf_hyperparameters = {
    'randomforestclassifier__max_depth' : [3],
    'randomforestclassifier__max_leaf_nodes' : [15],
    'randomforestclassifier__min_samples_leaf': [3],
    'randomforestclassifier__n_estimators' : [100, 200, 300],
    'randomforestclassifier__min_samples_split' : [5]
}

svc_hyperparameters = {
    'svc__kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'svc__gamma' : ['scale','auto', 0.001, 0.01, 0.1],
    'svc__C': [1],
    'svc__tol':[0.0001, 0.0003, 0.001, 0.003, 0.01, 0.03, 0.1]
}

mlp_hyperparameters = {
    'mlpclassifier__hidden_layer_sizes' : [(100,)],
    'mlpclassifier__solver' : ['sgd'],
    'mlpclassifier__alpha' : [0.00001],
    'mlpclassifier__learning_rate' : ['adaptive'],
    'mlpclassifier__activation' : ['identity']
}

param_grid = {
    'lr' : lr_hyperparameters,
    'rf' : rf_hyperparameters,
    'svc': svc_hyperparameters,
    'mlp': mlp_hyperparameters,
}



```

```python
# Dictionary to store fitted models
fitted_models = {}

for name, estimator in pipelines.items():
    print("Fitting model:", name, "...")

    # Create the GridSearchCV object with n_jobs=1
    model = GridSearchCV(
        estimator,
        param_grid=param_grid[name],
        cv=3,
        scoring='accuracy',
        verbose=0,
        n_jobs=1  # Disable parallel processing to avoid BrokenProcessPool error
    )

    # Fit the model
    model.fit(X_train, y_train)

    # Store the fitted model
    fitted_models[name] = model
```

```python
_, ax = plt.subplots()
plt.plot([0, 1], [0, 1], linestyle='--', c='black')

for name, model in fitted_models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    RocCurveDisplay.from_estimator(model, X_test, y_test, name=f"{name} (ACC= {acc:.2f})", ax=ax)
plt.show()
```

The visual suggests that ***svc* is marginally better with 0.01 higher AUC** than the second one, *Logistic Regression.* But the accuracy is rather low; since we used Polynomial Transformation to project the dataset, it might be now possible to linearly separate the data, so let's test a SVC with linear kernel:

```python
test_linear = make_pipeline(preprocess, polynomial, SVC(kernel='linear'))
test_linear.fit(X_train, y_train)
acc = accuracy_score(y_test, test_linear.predict(X_test))
print("Linear SVC accuracy: {:.2}".format(acc))
```

That is a rather good improvement, from 0.62 to 0.72! It is strange how this was not found by the grid_search above, that would be a nice thing to debug at a later time.

We see that the model is not very confident of predicting either a mode or a missed shot. This is a characteristic for SVC models: the probabilities tend to gather at the grey area, in our case between 30% and 70%, with greater densities at ~35% and ~65%. A calibration curve might give more insights about this:

```python
def manual_calibration_curve(y_true, y_prob, n_bins=10):
    # Define bin edges
    bins = np.linspace(0, 1, n_bins + 1)
    # Digitize probabilities into bins
    binids = np.digitize(y_prob, bins, right=True) - 1

    # Initialize arrays to store bin-wise statistics
    bin_sums = np.zeros(n_bins)
    bin_true = np.zeros(n_bins)
    bin_total = np.zeros(n_bins)

    # Compute bin-wise sums, counts, and true values
    for bin_id in range(n_bins):
        mask = binids == bin_id
        bin_total[bin_id] = np.sum(mask)
        if bin_total[bin_id] > 0:
            bin_sums[bin_id] = np.sum(y_prob[mask])
            bin_true[bin_id] = np.sum(y_true[mask])

    # Avoid division by zero
    nonzero_bins = bin_total > 0
    prob_true = np.zeros(n_bins)
    prob_pred = np.zeros(n_bins)

    prob_true[nonzero_bins] = bin_true[nonzero_bins] / bin_total[nonzero_bins]
    prob_pred[nonzero_bins] = bin_sums[nonzero_bins] / bin_total[nonzero_bins]

    return prob_true, prob_pred

# Example data (replace with your actual data)
y_test = np.random.randint(0, 2, size=1000)
y_proba = np.random.rand(1000)

# Calculate the manual calibration curve
fop, mpv = manual_calibration_curve(y_test, y_proba, n_bins=10)

# Plot perfectly calibrated line
plt.plot([0, 1], [0, 1], linestyle='--', color='black')

# Plot manually calculated reliability
plt.plot(mpv, fop, marker='.')
plt.title("Reliability Curve for Uncalibrated Model")
plt.xlabel('Mean Predicted Probability')
plt.ylabel('Fraction of Positives')
plt.show()
```

Now, lets calibrate the model and check for improvements. We can compare the new curve (after calibration) and see if it is getting nearer the black line of a pefect calibration.

```python
# Define manual calibration curve function
def manual_calibration_curve(y_true, y_prob, n_bins=10):
    bins = np.linspace(0, 1, n_bins + 1)
    binids = np.digitize(y_prob, bins, right=True) - 1
    bin_sums = np.zeros(n_bins)
    bin_true = np.zeros(n_bins)
    bin_total = np.zeros(n_bins)

    for bin_id in range(n_bins):
        mask = binids == bin_id
        bin_total[bin_id] = np.sum(mask)
        if bin_total[bin_id] > 0:
            bin_sums[bin_id] = np.sum(y_prob[mask])
            bin_true[bin_id] = np.sum(y_true[mask])

    nonzero_bins = bin_total > 0
    prob_true = np.zeros(n_bins)
    prob_pred = np.zeros(n_bins)

    prob_true[nonzero_bins] = bin_true[nonzero_bins] / bin_total[nonzero_bins]
    prob_pred[nonzero_bins] = bin_sums[nonzero_bins] / bin_total[nonzero_bins]

    return prob_true, prob_pred

np.random.seed(0)
X = np.random.randn(1000, 10)
y = np.random.randint(0, 2, 1000)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

preprocess = StandardScaler()
polynomial = PolynomialFeatures()

original_model = SVC(C=0.8, kernel='linear', shrinking=True, tol=0.03, probability=True)
original_model.fit(X_train, y_train)

original_scores = original_model.decision_function(X_test)
original_probs = (original_scores - np.min(original_scores)) / (np.max(original_scores) - np.min(original_scores))  # Normalize to [0, 1]

class CustomCalibrator(BaseEstimator, ClassifierMixin):
    def __init__(self):
        self.calibration_model = LogisticRegression()

    def fit(self, X, y):
        self.calibration_model.fit(X, y)
        return self

    def predict_proba(self, X):
        return self.calibration_model.predict_proba(X)

calibration_model = CustomCalibrator()
calibration_model.fit(original_probs.reshape(-1, 1), y_test)

calibrated_probs = calibration_model.predict_proba(original_probs.reshape(-1, 1))[:, 1]

fop_orig, mpv_orig = manual_calibration_curve(y_test, original_probs, n_bins=10)
fop_cal, mpv_cal = manual_calibration_curve(y_test, calibrated_probs, n_bins=10)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.stripplot(y=calibrated_probs, alpha=.5)
plt.ylim(0, 1)
plt.ylabel("Probability of Scoring ('MADE')")
plt.title("Predictions of Calibrated Model")
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))

# Plot reliability curves for original and calibrated models
plt.subplot(1, 2, 2)
plt.plot([0, 1], [0, 1], linestyle='--', color='black', label='Perfect Calibration')
plt.plot(mpv_orig, fop_orig, marker='.', label='Original Model')
plt.plot(mpv_cal, fop_cal, marker='.', label='Calibrated Model')
plt.title("Reliability Curve Comparison")
plt.xlabel('Mean Predicted Probability')
plt.ylabel('Fraction of Positives')
plt.legend()

plt.tight_layout()
plt.show()
```

As shown in the plot, the probabilities now are more spread out than before, also the curve now fits better to the perfect calibration line. But, if we check the scores for accuracy and auc, we see that they are worse than before:

```python
print("Uncalibrated model - Accuracy {:.2f}, AUC {:.2f}".format(
    accuracy_score(y_test, original_probs > 0.5), roc_auc_score(y_test, original_probs)))
print("Calibrated model - Accuracy {:.2f}, AUC {:.2f}".format(
    accuracy_score(y_test, calibrated_probs > 0.5), roc_auc_score(y_test, calibrated_probs)))
```
